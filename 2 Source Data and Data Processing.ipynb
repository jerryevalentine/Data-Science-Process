## Source Data ##

```python
import pandas as pd

source_data = "csv"
#source_data = "txt"
#source_data = "sql"

data_input_path = 'Resources/comic_books.csv'
data_output_path = 'Resources/comic_books_clean.csv'

if source_data == "csv":
    df = pd.read_csv(data_input_path)
    print("csv loaded from " + source_data)
elif source_data == "txt":
    print("txt loaded from " + source_data)
elif source_data == "sql":
    print("sql loaded from " + source_data)
```

# Previw Data #

```python
df.head()
```

# Change Column Names #

```python
# Your string representation of a dictionary
#a = "{'Old Name': 'New Name', 'Old Name': 'New Name'}"
a = "{'Title': 'Comic_Title', 'Date of publication': 'Publication_date'}"

# Convert the string to a dictionary using eval() - Note: use this cautiously
column_mapping = eval(a)

# Renaming columns
df.rename(columns=column_mapping, inplace=True)

# Displaying the DataFrame after renaming columns
print("\nDataFrame after renaming columns:")
df.head()
```

# Find Nulls #

```python


# Finding null values
null_values = df.isnull()

# Displaying the DataFrame with boolean values indicating null positions
print("\nDataFrame with Null Values (True indicates null):")


# Counting null values in each column
null_count = df.isnull().sum()

# Displaying the count of null values in each column
print("\nCount of Null Values in Each Column:")
print(null_count)
```

# Delete rows/columns #

```python

#reference activity 4.2.3 Cleaning_Data

#Delete rows with null values

df_cleaned = df.dropna()

# Displaying the DataFrame after dropping rows with null values
print("\nDataFrame after dropping rows with null values:")
df_cleaned.head()


## Delete extraneous column
del df['column_name']
## verify dropped columns/rows
df.count()

##verify data types
df.dtypes()
```

# Remove Special characters (non-numeric or non-alphabet) #

```python
# Creating a sample DataFrame with special characters
data = {'A': ['apple', '@banana', 'cherry!', '123%orange']}
data = pd.DataFrame(data)

# Displaying the original DataFrame
print("Original DataFrame:")
print(data)

# Removing special characters using the replace method with regex
data.replace(regex=r'[^a-zA-Z0-9\s]', value='', inplace=True)

# Displaying the DataFrame after removing special characters
print("\nDataFrame after removing special characters:")
print(data)
```

# Save Data #

```python
if source_data == "csv":
    # Write the DataFrame to a CSV file
    df.to_csv(data_output_path, index=False)

    print(f'csv data has been written to {data_output_path}')
elif source_data == "txt":
    print(f'txt data has been written to {data_output_path}')
elif source_data == "sql":
    print(f'sql data has been written to {data_output_path}')

```

```python
# Clean up values in a column (eg, 'Not employed' is the same as 'umemployed'. In this case, only one value is needed. Replace 'A' with 'B' in 'column'
df['column'] = df['column'].replace({'A': 'B'})
df['column'].value_counts()
```
